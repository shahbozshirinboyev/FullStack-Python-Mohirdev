# Nesting
Bu biror strukturani boshqasining ichida joylashtirish.

Pythonda nesting bu:


*   lug'atni ichida lug'at joylashishi
*   lug'atni ichida ro'yxat joylashishi
*   ro'yxatni ichida ro'yxat joylashishi
*   ro'yxatni ichida lug'at joylashishi

```python
car0 = {
    'model': 'Lacetti',
    'rang': 'oq',
    'yil': 2018,
    'narh': 13000,
    'km': 50000,
    'korobka': 'avtomat'
}

car1 = {
    'model': 'Nexia 3',
    'rang': 'kulrang',
    'yil': 2020,
    'narh': 10000,
    'km': 35000,
    'korobka': 'mexanika'
}

car2 = {
    'model': 'Onix',
    'rang': 'qora',
    'yil': 2023,
    'narh': 16000,
    'km': 10000,
    'korobka': 'avtomat'
}
```

```python
car = car0
print(f"{car['model'].title()}, {car['yil']}-yil, {car['km']}km yurgan, narxi {car['narh']}$")
car = car1
print(f"{car['model'].title()}, {car['yil']}-yil, {car['km']}km yurgan, narxi {car['narh']}$")
car = car2
print(f"{car['model'].title()}, {car['yil']}-yil, {car['km']}km yurgan, narxi {car['narh']}$")
```

**Nesting** - orqali osonroq yo'lini qilamiz:

```python
cars = [car0, car1, car2] #Nesting qilingan ro'yxat.
for car in cars:
  print(f"{car['model'].title()}, {car['yil']}-yil, {car['km']}km yurgan, narxi {car['narh']}$")
```

# While, Ro'yxatlar va Lug'atlar

```python
print("Do'stlaringiz ro'yxatini tuzamiz")
n = 1
dostlar = []
while True:
  savol = f"{n}-do'stingizni ismini kiriting: "
  dost = input(savol)
  dostlar.append(dost)
  n += 1
  takrorlash = input("Davom ettirasizmi? (ha/yo'q) ")
  if takrorlash != 'ha':
    break
for dost in dostlar:
  print(dost.title())
```

```python
print("Do'stlaringizni lug'atini yaratamiz")
dostlar = {}
n = 1
ishora  = True
while ishora:
  ism = input(f"{n}-do'stingizni ismini kiriting: ")
  yosh = input(f"{ism.title()}'ning yoshi nechada? ")
  dostlar[ism] = yosh
  n += 1
  davomi = input("Davom ettirasizmi? (ha/yo'q) ")
  if davomi != 'ha':
    ishora = False

for ism, yosh in dostlar.items():
  print(f"{ism.title()}'ning yoshi {yosh}'da")
```

```python
talabalar = ['botir', 'hasan', 'husan', 'fayyoza']
baholangan_talabalar = {}
while talabalar:
  talaba = talabalar.pop()
  baho = input(f"{talaba.title()}'ning bahosini kiriting: ")
  baholangan_talabalar[talaba] = int(baho)
  print(f"{talaba.title()} baholandi.")

print(baholangan_talabalar)
```
