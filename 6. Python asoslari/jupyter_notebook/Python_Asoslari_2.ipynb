# FOR SIKLI BILAN TANISHUV.
> for() operatori bilan ishlashni o'rganamiz!

```python
# Misol uchun har bir ismlar nomli ro'yxat elementi ichidagi ismlarga Salom, ism!
# deb yozib chiqishni automatik qilish uchun har for() ishlatilinadi.

mehmonlar = ['Ali', 'Vali', 'Hasan', 'Husan', 'Olim']
print('Salom,', mehmonlar[0])
print('Salom,', mehmonlar[1])
print('Salom,', mehmonlar[2])
print('Salom,', mehmonlar[3])
print('Salom,', mehmonlar[4])

# Yuqorida kodni qayta qayta takrorladik, bu ko'rinishdagi kodni qayta qayta takrorlamaslik uchun
# for() operatoridan foydalanamiz:
print("-----------------------------")

for mehmon in mehmonlar:
  print("Salom,", mehmon)

# for ichidagi mehmon o'zgaruvchisi ixtiyoriy belgilashimiz mumkin
# lekin tushunarliroq bo'lgani yaxshi
```

# BIR NECHTA SHARTLARNI TEKSHIRISH

**if-elif-else** ketma-ketligi haqida gaplashamiz:
Agar dastur tuzish davomida bir nechta shartni tekshirish kerak bo'lsa if-elif-else ketma-ketligidan foydalaniladi:
elif bu else va if so'zlarining jamlanmasi bo'lib, aks holda, agar deb tarjima qilinadi. Bunday if bilan boshlangan ketma-ketlik bir nechta elif lardan iborat bo'lishi mumkin.

```python
# Yoshga qarab narx belgilovchi dasturni tuzib chiqamiz:
yosh = int(input("Yoshingizni kiriting: >>> "))
if yosh <= 4:
  print("Sizga kirish bepul!")
elif yosh <= 12:
  print("Sizga kirish 5000 so'm!")
elif yosh <= 18:
  print("Sizga kirish 8000 so'm!")
else:
  print("Sizga kirish 10000  so'm!")

# Demak if-elif-else ketma ketligi yuqoridagidek ishlaydi va siz elif ni istalgancha ko'paytirishingiz mumkin.
```

```python
# Yuqoridagi dasturda ko'p marta print funksiyasi chaqirilgan, bu dasturni takrorlanishini
# ko'rsatib beradi, uni oldini olish uchun quyidagi yo'l tutish mumkin bo'ladi:
yosh = int(input("Yoshingizni kiriting: >>> "))
price = 0
if yosh <= 4:
  price = 0
elif yosh <= 12:
  price = 2000
elif yosh <= 18:
  price = 5000
else:
  price = 10000
print(f"Sizga kirish {price} so'm!")

# bunda kodni takrorlanishini oldi olinadi.
```

Yuqoridagi holatlarda turli xil ifodalarni solishtirishda ulardan qaytgan qiymat true yoki false bo'lishini ko'rdik. Bu qiymatlar boolean (mantiqiy) qiymatlar deb ataladi.

```python
# Pythonda o'zgaruvchilarni boolean qiymatda ham saqlash mumkin.
narx = 15000      # Mijoz 15.000 so'mga toam oldi.
choy = True       # Mijoz choy ham oldi.
salat = False     # Mijoz salat olmadi.

if choy and salat:
  narx = narx + 10000
elif choy or salat:
  narx = narx + 5000

print(f"Jami narx: {narx} so'm.")
```

```python
# Boolean o'zgaruvchilarni saqlashda True/False qiymatlarini o'rniga 1 va 0 dan foydalanish mumkin.
narx = 15000      # Mijoz 15.000 so'mga toam oldi.
choy = 1       # Mijoz choy ham oldi.
salat = 0     # Mijoz salat olmadi.

if choy and salat:
  narx = narx + 10000
elif choy or salat:
  narx = narx + 5000

print(f"Jami narx: {narx} so'm.")
```
